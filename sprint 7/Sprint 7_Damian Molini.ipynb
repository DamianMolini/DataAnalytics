### Exercici 1

##### Calculadora de l'índex de massa corporal

##### - Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.

##### - La funció ha de classificar el resultat en les seves respectives categories

```python
def calcular_IMC(peso, altura):
    """Calcular IMC y asignarle una categoría"""
    
    IMC = peso / (altura ** 2)

    if IMC < 18.5:
        categoria = "Bajo peso"
    elif IMC >= 18.5 and IMC < 25:
        categoria = "Peso normal"
    elif IMC >= 25 and IMC < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"

    return IMC, categoria

peso = float(input("Ingrese su peso en kg: "))
altura = float(input("Ingrese su altura en metros (p. ej.: 1.70) "))    

print(f"IMC: {calcular_IMC(peso, altura)[0]}")
print(f"Categoría: {calcular_IMC(peso, altura)[1]}")
```

### Exercici 3

##### Comptador de paraules d'un text.

##### Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
import re

def contar(texto):
    """Contar las veces que aparece cada palabra en un texto"""
    
    conteo = {}
    texto_limpio = re.sub(r'[^\w\s]', '', texto)
    texto_limpio = texto_limpio.lower()
    palabras = texto_limpio.split()
    
    for palabra in palabras:
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1
            
    return conteo

oracionPrueba = "What will you do? Ever tried, ever failed, no matter! Try again, fail again, fail better."

contar(oracionPrueba)
```

### Exercici 4

##### Diccionari invers.

##### Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.


### **Diccionario sin duplicados:**

```python
def invertir_diccionario(dic):
    """Invertir claves y valores de un diccionario. 
    Si los valores del dic original están duplicados, habrá un mensaje de advertencia"""
    
    igual_cantidad_valores = len(dic) == len(set(dic.values())) #esta variable es True si el número de items del dic es igual al número de valores únicos
    if igual_cantidad_valores == True:
        #si es true, todos los valores son únicos y podemos proceder a invertir el diccionario con un bucle
        dicInvertido = {valor: clave for clave, valor in dic.items()}
        return dicInvertido
    else:
        raise ValueError("No se puede invertir el diccionario porque existen valores repetidos.")

sin_duplicados = {"españa": "madrid", "italia": "roma", "francia": "parís"}
print(invertir_diccionario(sin_duplicados))
```

### Exercici 1

##### Diccionari invers amb duplicats

##### Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En tals casos, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que un valor únic no ha de ser una llista.

```python
def invertir_dic_duplicados(dic):
    """Invertir claves y valores de un diccionario. Si los valores del dic original están duplicados, 
    las claves del dic original se almacenarán como una lista de valores en el dic resultante"""
    
    try:
        invertir_diccionario(dic) #intentamos correr la función creada en el ejercicio anterior
        return dic
    except ValueError: #si da error, es porque existen valores duplicados
        dicInvertido = {}
        for clave, valor in dic.items():
            #si el valor del dic original NO existe en el dic invertido, lo agregamos (cambiando el orden de valor y clave)
            if valor not in dicInvertido:
                dicInvertido[valor] = clave            
            else:
                #si el valor del dic original SÍ existe en el dic invertido, hay dos posibilidades: 
                if isinstance(dicInvertido[valor], list):
                    #si el valor del dic invertido es una lista, significa que el valor del dic original ya se repitió antes, 
                    #por lo que debemos añadir su clave a la lista existente en el dic invertido
                    dicInvertido[valor].append(clave) 
                else:
                    #si no es una lista, significa que es la primera vez que se repite, y es aquí donde crearemos esa lista
                    dicInvertido[valor] = [dicInvertido[valor], clave]
        return dicInvertido 
    

sin_duplicados = {"españa": "madrid", "italia": "roma", "francia": "parís"}
con_duplicados = {"coche": "rojo", "camión": "blanco", "avión": "blanco", "tranvía": "negro", "bus": "blanco", "metro": "blanco"}
  
print(f"Sin duplicados: {invertir_dic_duplicados(sin_duplicados)}")
print(f"Con duplicados: {invertir_dic_duplicados(con_duplicados)}")
```

### Exercici 2

##### Conversió de tipus de dades

##### El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
def convertir_lista(lista):
    """dividir una lista en dos: 
    una con elementos que se puedan convertir a floats, y otra con los que no"""
    
    lista_con_floats = []
    lista_sin_floats = []
    for i in lista:
        try:
            lista_con_floats.append(float(i))
        except:
            lista_sin_floats.append(i)
    
    return lista_con_floats, lista_sin_floats

listaPrueba = [ "1.3", "one" , "1e10" , "seven", "3-1/2", ("2",1,1.4,"not-a-number"), [1,2,"3","3.4"]]

print(f"Lista con floats: {convertir_lista(listaPrueba)[0]}")
print(f"Lista sin floats: {convertir_lista(listaPrueba)[1]}")
```

### Exercici 1

##### Comptador i endreçador de paraules d'un text.

##### El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:



```python
import re

def contar(texto):

    """Contar las veces que aparece cada palabra en un texto y ordenarlas alfábeticamente 
    dentro de un diccionario cuya clave es la primera letra de cada palabra"""
    
    conteo = {}
    texto_limpio = re.sub(r'[^\w\s]', '', texto)
    texto_limpio = texto_limpio.lower()
    palabras = texto_limpio.split()
    
    for palabra in palabras:
        #definimos la primera letra de cada palabra
        primera_letra = palabra[0]
        #si la primera letra no está en la lista "conteo", la agregamos como clave y su valor es un diccionario vacío
        if primera_letra not in conteo:
            conteo[primera_letra] = {}
        #si la palabra ya existe en el diccionario de su primera letra, le sumamos 1, y si no existe, la agregamos con el valor 1
        if palabra in conteo[primera_letra]:
            conteo[primera_letra][palabra] += 1
        else:
            conteo[primera_letra][palabra] = 1

    #ordenamos las claves alfabéticamente
    conteo = dict(sorted(conteo.items()))
    
    #por último, para cada clave, ordenamos los valores alfabéticamente
    for letra, palabras in conteo.items():
        conteo[letra] = dict(sorted(palabras.items()))
    
    return conteo

with open("C:\\Users\\damim\\Desktop\\tu_me_quieres_blanca.txt", encoding='utf-8') as file:
    conteo_palabras = contar(file.read())

conteo_palabras
```
